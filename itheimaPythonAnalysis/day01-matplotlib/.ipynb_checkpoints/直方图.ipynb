```python
from matplotlib import pylab

#解决matplotlib显示中文的问题
pylab.rcParams['font.sans-serif'] = ['SimHei']

# 1、准备数据
time_data = [131,98,125,131,124,139,131,117,128,138,135,138,131,182,187,114,136,117,128,188,135,138,131,102,107,114,119,128,121,142,127,138,181,127,130,124,181,131,98,125,131,124,139,131,117,128,138,135,138,131,182,187,114,136,117,128,188,135,138,131,102,107,114,119,128,121,142,127,138,181,127,130,124,181]
# 2、创建画布
pylab.figure(figsize = (20,8),dpi = 80)
# 3、绘制直方图
bins = int((max(time_data)-min(time_data))/2)
pylab.hist(time_data,bins = bins)

# 修改x轴刻度
pylab.xticks(range(min(time_data),max(time_data),2))
pylab.title("电影时长的分布情况")

# 添加网格
pylab.grid(linestyle = "--",alpha = 0.5)

# 4、显示图像
pylab.show()
```
