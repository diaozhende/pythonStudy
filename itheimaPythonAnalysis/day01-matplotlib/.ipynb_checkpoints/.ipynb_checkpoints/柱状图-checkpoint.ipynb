### 柱状图绘制

```python
#对比电影票房收入

from matplotlib import pylab

#解决matplotlib显示中文的问题
pylab.rcParams['font.sans-serif'] = ['SimHei']

# 1、准备数据
movel_name = ["雷神三","红海行动","正义联盟","寻梦环游记","金秋风暴","降魔传","追捕","战狼2","西虹市首富","极速人生","鼠胆英雄"]
movel_count = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6754,52222]

# 2、创建画布
pylab.figure(figsize = (20,8),dpi = 80)

# 3、绘制柱状图 color:指定颜色
pylab.bar(range(len(movel_name)),movel_count)

# 修改x的刻度
pylab.xticks(range(len(movel_name)),movel_name)
# pylab.xticks(range(len(movel_name)))

# 添加网格
pylab.grid(linestyle="--",alpha=0.5)

# 显示图例
# pylab.legend()

# 4、显示图像
pylab.show()
```

```python
# 比较相同天数的票房

from matplotlib import pylab

#解决matplotlib显示中文的问题
pylab.rcParams['font.sans-serif'] = ['SimHei']

# 1、准备数据
movel_name = ["雷神三","红海行动","正义联盟"]
first_day = [10587,10062,1275]
first_weekend = [36224,34479,11830]

# 2、创建画布
pylab.figure(figsize = (20,8),dpi = 80)

# 3、绘制柱状图
pylab.bar(range(len(movel_name)),first_day,width=0.2)
# width：设置柱状图的宽度
pylab.bar([i+0.2 for i in range(len(movel_name))],first_weekend,width=0.2)

# 4、显示图像
pylab.show()
```
