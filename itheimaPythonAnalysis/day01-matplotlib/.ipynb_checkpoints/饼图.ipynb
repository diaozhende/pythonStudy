### 绘制饼图

```python
from matplotlib import pylab

#解决matplotlib显示中文的问题
pylab.rcParams['font.sans-serif'] = ['SimHei']

# 1、准备数据
movel_name = ["雷神三","红海行动","正义联盟","寻梦环游记","金秋风暴"]
movel_count = [73853,57767,22354,15969,14839]
# 2、创建画布
pylab.figure(figsize = (20,8),dpi = 80)

# 3、绘制饼图
pylab.pie(movel_count,labels=movel_name,autopct="%1.2f%%")

# 添加axis：为了让显示的饼图保持圆形，需要添加axis保持长宽一样
pylab.axis("equal")

# 4、显示图像
pylab.show()
```
