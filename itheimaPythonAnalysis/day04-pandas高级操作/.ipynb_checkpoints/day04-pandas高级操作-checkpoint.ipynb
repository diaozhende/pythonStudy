### 案例：股票的涨跌幅离散化

```python
# 将p_change进行离散化
```

```python
# 1、读取数据
data = pandas.read_csv("E:/python大数据资料/黑马-01-Python数据挖掘基础资料/资料/day4资料/02-代码/stock_day.csv")
```

```python
# 2、对数据进行自动分组
stockData = data["p_change"]
stock_qcut = pandas.qcut(stockData,10)
```

```python
# 3、将数据转换成one-hot编码
pandas.get_dummies(stock_qcut,prefix="涨跌幅")
```

```python
# 手动分组
stock_cut = pandas.cut(stockData,[-100, -7, -5, -3, 0, 3, 5, 7, 100])
stock_cut_one_hot = pandas.get_dummies(stock_cut,prefix="涨跌幅")
stock_cut_one_hot.head()
```

### 3.pandas.concat()按方向进行合并

```python
# pandas.concat([data1,data2],axis=1) 当axis=1时是竖直拼接，当axis=0时是水平拼接
```

```python
# 处理好的one-hot编码与原数据合并
# data,stock_cut_one_hot
pandas.concat([data,stock_cut_one_hot],axis=1)
```

```python
pandas.concat([data,stock_cut_one_hot],axis=0)
```

### 4.pandas.merge() 按键进行合并

```python
# pandas.merge(left,right,how='',on=[])
# left,right：左表和右表
# how：合并方式(left:左连接，right：右连接，inner：内连接，outer：外连接)
# on:按照哪些键进行合并
```

```python
left = pandas.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pandas.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

### 4.1 内连接

```python
# 内连接：将两个表中相同的数据进行合并
pandas.merge(left,right,how="inner",on=["key1","key2"])
```

### 4.2 左连接

```python
# 左连接：以左表为基准，和右表数据进行合并，如果左表中的数据右表没有，那么右表中的键为NaN
pandas.merge(left,right,how="left",on=["key1","key2"])
```

### 4.3 右连接

```python
# 右连接：以右表为基准，和左表数据进行合并，如果右表中的数据左表没有，那么左表中的键为NaN
pandas.merge(left,right,how="right",on=["key1","key2"])
```

### 4.4 外连接

```python
# 外连接：将两个表的所有数据进行合并，那个表中没有这条数据，那么键显示NaN
pandas.merge(left,right,how="outer",on=["key1","key2"])
```

### 案例：探求股票涨跌幅与星期几的关系

```python
stock_Data = pandas.read_csv("E:/python大数据资料/黑马-01-Python数据挖掘基础资料/资料/day4资料/02-代码/stock_day.csv")
```

```python
stock_Data.index
```

```python
# 将字符串的日期转换成pandas日期类型
date = pandas.to_datetime(stock_Data.index)
```

```python
# 获取日期是星期几
date.weekday
```

```python
# 将星期几添加到stock_data列中
stock_Data["week"] = date.weekday
```

```python
stock_Data.head()
```

```python
# 准备涨跌幅数据
p_change_Data = numpy.where(stock_Data["p_change"]>0,1,0)
# 将涨跌幅数据添加到stock_data列中
stock_Data["p_change_Data"] = p_change_Data
```

```python
stock_Data
```

```python
# 交叉表
tab_data = pandas.crosstab(stock_Data["week"],stock_Data["p_change_Data"])
```

```python
tab_data
```

```python
tab_data.sum(axis=1) # 当axis=1时是每行求和，当axis=0时是每列求和
```

```python
# 计算比率
tab_data.div(tab_data.sum(axis=1),axis=0)
```

```python
# 画柱状图
tab_data.div(tab_data.sum(axis=1),axis=0).plot(kind="bar")
```

```python
tab_data.div(tab_data.sum(axis=1),axis=0).plot(kind="bar",stacked=True)
```

### 透视表

```python
stock_Data.pivot_table(["p_change_Data"],index = ["week"])# 安装week进行分组计算比率
```

### 3.问题3：对于这一组电影数据，如果我们希望统计电影分类(genre)的情况，应该如何处理数据？

```python
movieData.head()
```

```python
genreData1 = [i.split(",") for i in movieData["Genre"]]
genreData1
```

```python
genreData2 = [j for i in genreData1 for j in i]
```

```python
movie_class = numpy.unique(genreData2)
```

```python
# 统计每个类别中有多少个电影
# 生成一个dataFram进行站位
data = pandas.DataFrame(numpy.zeros(shape=[len(movieData["Title"]),movie_class.size],dtype=numpy.int32),columns=movie_class)
```

```python
# 计数填表
for i in range(len(data)):
    data.ix[i,genreData1[i]] = 1
```
