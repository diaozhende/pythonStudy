### 案例：将股票的涨跌幅数据进行离散化

```python
# 1、读取数据
gpData = pandas.read_csv("E:/python大数据资料/黑马-01-Python数据挖掘基础资料/资料/day4资料/02-代码/stock_day.csv")
gpData.head()
```

```python
# 2、分组
p_change = gpData["p_change"]
groupData = pandas.qcut(p_change,10)
groupData.value_counts()
```

```python
# 将数据转换成one-hot编码
change_data = pandas.get_dummies(groupData,prefix="涨跌幅")
```

### 6、表的合并

#### (1)concat()

```python
# 处理好的one-hot编码与原表进行合并
gpData.head()
```

```python
change_data.head()
```

```python
pandas.concat([gpData,change_data],axis=1).head() # axis=1代表水平合并，=0代表竖直合并
```

#### (2)merge()

```python
left = pandas.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pandas.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pandas.merge(left,right,how="left",on=["key1","key2"])
```

```python
pandas.merge(left,right,how="inner",on=["key1","key2"])
```

```python
pandas.merge(left,right,how="outer",on=["key1","key2"])
```

### 7、crosstab(交叉表)

```python
# 使用交叉表统计涨跌的股票数量
```

```python
# 1、准备数据
gpData = pandas.read_csv("E:/python大数据资料/黑马-01-Python数据挖掘基础资料/资料/day4资料/02-代码/stock_day.csv")
gpData.head()
```

```python
# 2、准备星期数据列
date = pandas.to_datetime(gpData["Unnamed: 0"].values)
```

```python
gpData["week"] = date.weekday
```

```python
# 4、交叉表
data = pandas.crosstab(gpData["week"],gpData["pona"])
```

```python
data
```

```python
data.div(data.sum(axis=1),axis=0).plot(kind="bar",stacked=True)
```

#### 问题2：对于这一组电影数据，如果我们想rating，runtime的分布情况，应该如何呈现数据？

```python
# 展示Rating的分布情况
import matplotlib.pylab
# 1、创建画布
matplotlib.pylab.figure(figsize=(20,8),dpi=80)
# 2、画直方图
matplotlib.pylab.hist(movieData["Rating"],20)
# 修改刻度
matplotlib.pylab.xticks(numpy.linspace(movieData["Rating"].min(),movieData["Rating"].max(),21))
# 添加网格
matplotlib.pylab.grid(linestyle="--",alpha=0.5)
# 3、显示图像
matplotlib.pylab.show()
```

#### 问题3：对于这一组电影数据，如果我们希望统计电影分类(genre)的情况，应该如何处理数据？

```python
movie_genre_data = [i.split(",")for i in movieData["Genre"]]
```

```python
movie_genre_data 
```

```python
movie_class_data = numpy.unique([j for i in movie_genre_data for j in i])
```

```python
movie_class_data.size
```

```python
movieData.size
```

```python
count = pandas.DataFrame(numpy.zeros(shape=[len(movie_genre_data),movie_class_data.size],dtype="int32"),columns=movie_class_data)
```

```python
len(count)
```

```python
for i in range(len(count)):
    count.ix[i,movie_genre_data[i]] = 1
```

```python
count.head()
```

```python
count.sum(axis=0).sort_values(ascending=False).plot(kind="bar",figsize=(20,9),fontsize=40,colormap="cool")
```
