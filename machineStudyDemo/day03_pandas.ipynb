### 1.DataFrame

### 1.1DataFrame属性

```python
# 创建一个符合正态分布的10个股票5天的涨跌幅
import numpy
gpData = numpy.random.normal(0,1,(10,5))
```

```python
# 将array数组转换成DataFrame
import pandas
gpFrame = pandas.DataFrame(gpData)
```

```python
# shape:DataFram的形状（行，列）
gpFrame.shape
```

```python
# index：DataFrame的行索引
# 添加行索引
indexData = ["股票{}".format(i) for i in range(10)]
gpFrame = pandas.DataFrame(gpData,index=indexData)
```

```python
gpFrame
```

```python
# columns:列索引
# 添加列索引
date = pandas.date_range(start="20180101",periods=5,freq="B")
gpFrame = pandas.DataFrame(gpData,index=indexData,columns=date)
gpFrame
```

```python
# values:DateFrame中的数据，也就是array数组
gpFrame.values
```

```python
# T:将DataFrame进行转置（行变列，裂变行）
gpFrame.T
```

### 5、DataFrame运算

### 5.1、算数运算

```python
# add()：加
data["open"].add(100).head()
```

```python
# sub()：减
data["open"].sub(10).head()
```

```python
data["open"]*100# 也可以用运算符
```

### 5.2、逻辑运算

### 5.2.1、逻辑运算（<,>,|,&）

```python
# 例如筛选p_change > 2的日期数据
data[data["p_change"]>2].head()
```

```python
data["p_change"]>2
```

```python
# 完成一个多个逻辑判断， 筛选p_change > 2并且low > 15
data[(data["p_change"]>2) & (data["low"]>15)].head()
```

### 5.2.2、逻辑运算函数

```python
# query()
data.query("p_change > 2 & low > 15").head()
```

```python
# isin()
# 判断'turnover'是否为4.19, 2.39
data[data["turnover"].isin([4.19, 2.39])]
```

### 5.4、统计运算

```python
# 累计统计函数
# cumsum：计算前n个数的和
# cummax：计算前n个数的最大值
# cummin：计算前n个数的最小值
# cumprod：计算前n个数的积
```
