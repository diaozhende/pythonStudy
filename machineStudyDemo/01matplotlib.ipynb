# 画出城市某一时刻的温度变化

```python
from matplotlib import pylab
import matplotlib.pyplot
matplotlib.pyplot.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
import random
# 创建数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
pylab.figure(figsize=(20,8),dpi=80)
pylab.plot(x,y_shanghai,color="m",linestyle="--",label="shanghai")
# y轴0-40，步长为5
pylab.yticks(range(40)[::5])
xLabel = ["11:{}".format(i) for i in x]
pylab.xticks(x[::5],xLabel[::5])
# 增加网格
pylab.grid(linestyle="--",alpha = 0.5)
# 设置标题
# pylab.xlabel("时间")
# pylab.ylabel("温度")
# pylab.title("上海十一点到十二点的温度变化图")

# 添加第二条折线
y_beijing = [random.uniform(2,5) for i in x]
pylab.yticks(range(0,40,5))
pylab.plot(x,y_beijing,color="g",linestyle="-",label="beijing")
# 添加图例
pylab.legend("upper right")
matplotlib.pylab.xlabel("时间")
matplotlib.pylab.title("上海和北京11点-12点的温度变化图")
matplotlib.pylab.ylabel("温度")
pylab.show()
```

### 在一个画布中画两个两个城市温度变化图

```python
import matplotlibotl
import random
# 准备数据
x = range(0,60)
# random.uniform:生成一个区间内均匀的数组
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(2,5) for i in x]

# 创建多个子图
figure,axes = pylab.subplots(nrows=1,ncols=2,figsize=(20,8),dpi= 80)
# 绘制图像
axes[0].plot(x,y_shanghai,color = "m",linestyle="--",label="上海")
axes[1].plot(x,y_beijing,color="g",label = "北京")
# 显示图例
axes[0].legend()
axes[1].legend()
# 修改刻度
x_label = ["11:{}".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label)
axes[0].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label)
axes[1].set_yticks(range(0,40,5))
# 添加网格
axes[0].grid(linestyle="--",alpha=0.5)
axes[1].grid(linestyle="--",alpha=0.5)

# 添加描述信息
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")
pylab.show()
```
